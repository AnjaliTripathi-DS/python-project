# Visualization

```python
plt.hist(df['lead_time'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Lead Time')
plt.ylabel('Frequency')
plt.title('Lead Time Distribution')
plt.show()
```

This creates a histogram that shows the distribution of lead times, helping to understand how far in advance customers typically book their hotel stays. The x-axis represents lead time values, the y-axis represents the frequency (count) of lead times falling within each bin, and each bar in the histogram represents a bin of lead time values.

```python
segment_counts = df['market_segment'].value_counts()
plt.bar(segment_counts.index, segment_counts.values, color='lightcoral')
plt.xlabel('Market Segment')
plt.ylabel('Count')
plt.title('Distribution of Market Segments')
plt.xticks(rotation=45)
plt.show()
```

This creates a bar chart that visualizes the distribution of different market segments in the dataset. Each bar represents a market segment, and the height of the bar indicates the count of reservations associated with that segment. It helps understand the composition of customers based on their market segment categories.

```python
deposit_counts = df['deposit_type'].value_counts()
plt.pie(deposit_counts, labels=deposit_counts.index, autopct='%1.1f%%', colors=['pink', 'lightgreen', 'lightblue'])
plt.title('Distribution of Deposit Types')
plt.show()
```

This creates a pie chart that visualizes the distribution of different deposit types in the dataset. Each slice of the pie represents a deposit type, and the size of each slice corresponds to the percentage of reservations associated with that deposit type. The percentage labels on each slice provide additional information about the proportion of each deposit type in the dataset.
